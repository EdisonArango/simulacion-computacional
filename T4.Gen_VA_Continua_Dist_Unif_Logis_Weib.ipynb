```python
%matplotlib inline
import math   
import matplotlib.pyplot as plt    

def inverselogistic(n, miu, s, x0):  
    values = []
    for i in range(n):
        xi = (170*x0 + 1)%30323
        ui = xi/30323.0
        gen = miu + (s * (math.log(ui/(1-ui))))         
        values.append(gen)
        x0 = xi    
    return values

def inverseuniform(n, a, b, x0):  
    values = []
    for i in range(n):
        xi = (170*x0 + 1)%30323
        ui = xi/30323.0  
        gen = a + ((b - a)* ui)
        values.append(gen)
        x0 = xi
    return values

def inverseweibull(n, _lambda, k, x0): 
    values = []
    for i in range(n):
        xi = (170*x0 + 1)%30323
        ui = xi/30323.0
        gen = _lambda * (-math.log(1 - ui) )** k
        values.append(gen)
        x0 = xi     
    return values
```

# Logistic Distribution

```python
x = inverselogistic(100, 10, 2, 7)
plt.hist(x)
print x
```

# Uniform Distribution

```python
x = inverseuniform(100, 5, 10, 7)
plt.hist(x)
print x
```

# Weibull Distribution

```python
x = inverseweibull(100, 1, 3, 7)
plt.hist(x)
print x
```
