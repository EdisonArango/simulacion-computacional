```python
%matplotlib inline
import math   
import matplotlib.pyplot as plt    

def inverseexponential(n, _lambda, x0):
    values = []
    for i in range(n):
        xi = (170*x0 + 1)%30323
        ui = xi/30323.0
        gen = - (math.log(1 - ui))/_lambda
        values.append(gen)
        x0 = xi        
    return values

res = inverseexponential(100, 10, 7)
plt.hist(res)
print "Generated: ",res
```
