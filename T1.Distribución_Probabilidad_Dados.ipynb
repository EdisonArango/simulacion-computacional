```python
%matplotlib inline
import collections as cl
import matplotlib.pyplot as plt
from decimal import *

dice8f = range(1,9)
dice13f = range(1,14)
sums = []
for d1 in dice8f:
    for d2 in dice13f:
        sums.append(d1 + d2)
sortedCounts = cl.Counter(sums).most_common()
sortedCounts.sort()

getcontext().prec = 5

amount = len(sums)
prob = []
for i in sortedCounts:
    prob.append([i[0], i[1] / Decimal(amount)])

expected_value = sum([ x[0] * x[1] for x in prob ])
print "Expected value: {}".format(expected_value)
variance = sum([ (x[0] - expected_value)**2 * x[1] for x in prob ])
print "Variance: {}".format(variance)

plt.figure()
x =[[i[0]] for i in sortedCounts] 
y =[[i[1]] for i in prob] 
plt.plot(x,y)
plt.show()
```
